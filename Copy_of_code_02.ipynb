# Introdução

Como vimos na aula passada, uma imagem pode ser preto e branca, tons de cinza ou colorida. Ainda, conseguimos extrair informações das imagens através dos histogramas.

Para esta atividade, utilizaremos três imagens coloridas, retiradas do unsplash:

* johannes-plenio-RwHv7LgeC7s-unsplash.jpg

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import array
from PIL import Image
```

# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
path = '/content/drive/My Drive/Colab Notebooks/'

imagem_1 = Image.open(path + 'claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_2 = Image.open(path + 'bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem_3 = Image.open(path + 'johannes-plenio-RwHv7LgeC7s-unsplash.jpg')
imagem_1_arr = np.array(imagem_1)
imagem_2_arr = np.array(imagem_2)
imagem_3_arr = np.array(imagem_3)

plt.figure(figsize=(12,5))

plt.subplot(131)
plt.axis('off')
plt.imshow(imagem_1)

plt.subplot(132)
plt.axis('off')
plt.imshow(imagem_2)

plt.subplot(133)
plt.axis('off')
plt.imshow(imagem_3)

plt.show()

##################################################################

plt.figure(figsize=(12,5))

imagem_1_red = imagem_1_arr[:,:,0]
plt.subplot(2,3,1)
plt.hist(imagem_1_red.ravel(), color='red')

imagem_1_green = imagem_1_arr[:,:,1]
plt.subplot(2,3,2)
plt.hist(imagem_1_green.ravel(), color='green')

imagem_1_blue = imagem_1_arr[:,:,2]
plt.subplot(2,3,3)
plt.hist(imagem_1_green.ravel(), color='blue')
##################################################################

plt.figure(figsize=(12,5))

imagem_2_red = imagem_2_arr[:,:,0]
plt.subplot(3,3,1)
plt.hist(imagem_2_red.ravel(), color='red')

imagem_2_green = imagem_2_arr[:,:,1]
plt.subplot(3,3,2)
plt.hist(imagem_2_green.ravel(), color='green')

imagem_2_blue = imagem_2_arr[:,:,2]
plt.subplot(3,3,3)
plt.hist(imagem_2_blue.ravel(), color='blue')

###################################################################

plt.figure(figsize=(12,5))

imagem_3_red = imagem_3_arr[:,:,0]
plt.subplot(4,3,1)
plt.hist(imagem_3_red.ravel(), color='red')

imagem_3_green = imagem_2_arr[:,:,1]
plt.subplot(4,3,2)
plt.hist(imagem_3_green.ravel(), color='green')

imagem_3_blue = imagem_3_arr[:,:,2]
plt.subplot(4,3,3)
plt.hist(imagem_3_blue.ravel(), color='blue')


plt.show()

# imagem_color_arr = np.array(imagem_color)
```

Vamos gerar algumas imagens preto e branco, utilizando limiares diferentes:

*  45
*  100
*  145
*  200
*  245

Escreva o código no espaço abaixo.

```
img_1_arr = np.array(imagem_1)
plt.figure(figsize=(23,9))

img_45_gray = np.mean(img_1_arr, axis=2)
img_100_gray = np.mean(img_1_arr, axis=2)
img_145_gray = np.mean(img_1_arr, axis=2)
img_200_gray = np.mean(img_1_arr, axis=2)
img_245_gray = np.mean(img_1_arr, axis=2)

img_45_gray[img_45_gray > 45] = 45
img_45_gray[img_45_gray < 45] = 0
plt.subplot(151)
plt.axis('off')
plt.imshow(Image.fromarray(img_45_gray))

img_100_gray[img_100_gray > 100] = 100
img_100_gray[img_100_gray < 100] = 0
plt.subplot(152)
plt.axis('off')
plt.imshow(Image.fromarray(img_100_gray))

img_145_gray[img_145_gray > 145] = 145
img_145_gray[img_145_gray < 145] = 0
plt.subplot(153)
plt.axis('off')
plt.imshow(Image.fromarray(img_145_gray))

img_200_gray[img_200_gray > 200] = 200
img_200_gray[img_200_gray < 200] = 0
plt.subplot(154)
plt.axis('off')
plt.imshow(Image.fromarray(img_200_gray))

img_245_gray[img_245_gray > 245] = 245
img_245_gray[img_245_gray < 245] = 0
plt.subplot(155)
plt.axis('off')
plt.imshow(Image.fromarray(img_245_gray))

plt.show()
```
