# 1a parte

No espaço de código abaixo, plote as 3 imagens, uma ao lado da outra. Ainda, plote o histograma de cada canal de cor (R - Vermelho, G - Verde e B - Azul) abaixo da respectiva imagem.

```
path = '/content/drive/My Drive/Colab Notebooks/'

imagem_1 = Image.open(path + 'claudio-testa--SO3JtE3gZo-unsplash.jpg')
imagem_2 = Image.open(path + 'bailey-zindel-NRQV-hBF10M-unsplash.jpg')
imagem_3 = Image.open(path + 'johannes-plenio-RwHv7LgeC7s-unsplash.jpg')
imagem_1_arr = np.array(imagem_1)
imagem_2_arr = np.array(imagem_2)
imagem_3_arr = np.array(imagem_3)

plt.figure(figsize=(12,5))

plt.subplot(131)
plt.axis('off')
plt.imshow(imagem_1)

plt.subplot(132)
plt.axis('off')
plt.imshow(imagem_2)

plt.subplot(133)
plt.axis('off')
plt.imshow(imagem_3)

plt.show()

##################################################################

plt.figure(figsize=(12,5))

imagem_1_red = imagem_1_arr[:,:,0]
plt.subplot(2,3,1)
plt.hist(imagem_1_red.ravel(), color='red')

imagem_1_green = imagem_1_arr[:,:,1]
plt.subplot(2,3,2)
plt.hist(imagem_1_green.ravel(), color='green')

imagem_1_blue = imagem_1_arr[:,:,2]
plt.subplot(2,3,3)
plt.hist(imagem_1_blue.ravel(), color='blue')

##################################################################

plt.figure(figsize=(12,5))

imagem_2_red = imagem_2_arr[:,:,0]
plt.subplot(3,3,1)
plt.hist(imagem_2_red.ravel(), color='red')

imagem_2_green = imagem_2_arr[:,:,1]
plt.subplot(3,3,2)
plt.hist(imagem_2_green.ravel(), color='green')

imagem_2_blue = imagem_2_arr[:,:,2]
plt.subplot(3,3,3)
plt.hist(imagem_2_blue.ravel(), color='blue')

###################################################################

plt.figure(figsize=(12,5))

imagem_3_red = imagem_3_arr[:,:,0]
plt.subplot(4,3,1)
plt.hist(imagem_3_red.ravel(), color='red')

imagem_3_green = imagem_3_arr[:,:,1]
plt.subplot(4,3,2)
plt.hist(imagem_3_green.ravel(), color='green')

imagem_3_blue = imagem_3_arr[:,:,2]
plt.subplot(4,3,3)
plt.hist(imagem_3_blue.ravel(), color='blue')


plt.show()

# imagem_color_arr = np.array(imagem_color)
```
